```python
from langchain_community.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import transformers
```

```python
import torch
```

```python
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
model = "daryl149/llama-2-7b-chat-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.bfloat16,
    trust_remote_code = True,
    device_map = "auto",
    max_length = 1000,
    do_sample = True,
    top_k = 10,
    num_return_sequences = 1,
    eos_token_id = tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs={'temperature':0})
```

```python
prompt = "What would be a good name for a company that makes colorful socks"
```

```python
print(llm(prompt))
```

```python
prompt = "I want to open a restaurant for Indian food. Suggest me a fancy name for this"
```

```python
print(llm(prompt))
```

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
prompt_template1 = PromptTemplate(input_variables=["cuisine"], template = "I want to open a restaurant for {cuisine} food. Suggest a fancy name for this")
```

```python
input_prompt = prompt_template1.format(cuisine="indian")
```

```python
print(input_prompt)
```

```python
prompt_template2 = PromptTemplate(input_variables=["book_name"], template = "Provide me a concise summary of the book {book_name}")
```

```python
input_prompt = prompt_template2.format(book_name="Alchemist")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template1, verbose = True)
```

```python
response = chain.run("indian")
print(response)
```
